Exercício 4 - Calculadora de Desconto. 💵 🤑

Uma loja dá 10% de desconto para compras acima de R$ 100,00. Faça um programa que:

1º - Pergunta o valor da compra.

2º - Calcula o valor final com desconto (se aplicável).

3º - Mostra o valor a pagar.

```python
### Seu código aqui

valor = float(input("Qual o valor da compra?"))
if valor >100:
    desconto = valor * 0.10
    print(f"você fez {valor} em compras, teve o desconto de {desconto}, total a pagar {valor-desconto}")
else:
    print("Você não teve desconto")
```

Exercício 5 - Classificação de Notas 🧑‍🎓

Pergunte a nota de um aluno e dependendo da nota que ele informar, o classifique de acordo com as regras abaixo:

Classifique a nota do aluno:

- 9 até 10: "Excelente"

- 7 até 8.9: "Bom"

- 5 até 6.9: "Regular"

- 0 até 4.9: "Reprovado"

mostre para o aluno a classificação dele através de um print, exemplo print("Excelente").

```python
nota = float(input("Qual a sua nota ? :"))
if 9 <= nota <= 10:
    print("Excelente")
elif 7 <= nota <= 8.9:
    print("Bom")
elif 5 <= nota <= 6.9:
    print("Regular")
elif 0 <= nota <= 4.9:
    print("Reprovado")
else:
    print("Nota inválida")
```

Exercício 8 - Dia da Semana
Peça um número de 1 a 7 e mostre o dia da semana correspondente: 📅

- 1: Domingo

- 2: Segunda

- ... até 7: Sábado

```python
numero = int(input("Digite um número de 1 até 7"))

if numero == 1:
    print("Domingo")
elif numero == 2:
    print("Segunda-feira")
elif numero == 3:
    print("Terça-feira")
elif numero == 4:
    print("Quarta-feira")
elif numero == 5:
    print("Quinta-feira")
elif numero == 6:
    print("Sexta-feira")
elif numero == 7:
    print("Sábado")
else:
    print("Número inválido! Digite um valor de 1 a 7.")
```

Exercício 9 - Calculadora Simples 📏📐

- Faça uma calculadora que:

- Pede dois números

- Pede a operação (+, -, *, /)

- Dependendo da operação mostre o resultado.



```python
n1 = float(input("Primeiro número: "))
n2 = float(input("Segundo número: "))
op = input("Operação (+, -, *, /): ")

if op == "+":
    print(n1 + n2)
elif op == "-":
    print(n1 - n2)
elif op == "*":
    print(n1 * n2)
elif op == "/":
    print(n1 / n2)
```

Exercício 10 - Sistema de Multas 🚙🚘🧾


- Velocidade máxima: 80 km/h 🚫

- Até 80: "Sem multa"

- 81-100: "Multa leve"

- Acima de 100: "Multa grave"

```python
velocidade = int(input("Digite a velocidade: "))

if velocidade <= 80:
    print("Sem multa")
elif velocidade <= 100:
    print("Multa leve")
else:
    print("Multa grave")
```
